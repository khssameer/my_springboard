```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataFile = "Data/GoldPriceData.csv"

goldRateDf = pd.read_csv(dataFile, index_col='Date',parse_dates=True)
goldRateDf.head()
```

```python
print(goldRateDf.info())

fig, axes = plt.subplots(nrows=3, ncols=1)

print(axes)

goldRateDf.Open.plot(figsize=(10,5), color='red',ax=axes[0])
goldRateDf.High.plot(figsize=(10,5), color='blue',ax=axes[1])
goldRateDf.Close.plot(figsize=(10,5), color='green',ax=axes[2])

plt.show()
```
